```python
# investigating max speed as '240' seemed suspicious. Turns out its 240 kph. Formula Rossa in Abu Dhabi is fastest coaster in my data set.

top_speed = df[df["speed1_value"] == 240]
top_speed
```

note :
- Speed1_value and speed2_value are coaster's speed as float, speed1_unit speed2_unit explain what unit of measurement that speed is measured in.
- Same goes for height1 height2



```python
# investigation min height, again as 4 seemed suspicious. 13 feet height for kiddie coaster, wikipedia listing confirmed this is correct.

min_height = df[df["height_value"] == 4]
min_height
```

```python
# operational coasters

operating_coasters = df[df["Status"] == "Operating"]
operating_coasters
```

```python
#checking to find roller coasters at my home park, Six Flags St Louis

six_flags_stl = df[df["Location"].str.contains("Six Flags St. Louis", case=False, na=False)]
six_flags_stl
```

```python
# Roller coasters at a Disney park!

# Checks rows that contain the word "Disney" in the Location column. Not case sensitive and not null.

# Checked Disney parks and St. Louis because I am most familiar with these locations. 
# My familiarity with their details allows me to better gauge the accuracy of the dataset.
# One issue I noticed is that rides with name changes (e.g., California Screamin' -> Incredicoaster) 
# are treated as two separate entries. Additionally, the old version is incorrectly listed as operational.


disney = df[df["Location"].str.contains("Disney", case=False, na=False)]
disney
```
